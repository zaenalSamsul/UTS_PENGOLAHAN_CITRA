```python
from google.colab import files
uploaded = files.upload()  # Mengunggah file gambar

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fungsi utama
def main():
    # Membaca citra grayscale
    image_path = next(iter(uploaded.keys()))  # Mendapatkan nama file yang diunggah
    image = cv2.imread(image_path)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if image is None:
        print("Citra tidak ditemukan. Periksa kembali path citra Anda!")
        return

    # Menampilkan citra asli
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Citra Grayscale')
    plt.axis('off')

    # Menghitung histogram
    histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

    # Menampilkan histogram
    plt.subplot(1, 3, 2)
    plt.plot(histogram, color='black')
    plt.title('Histogram Citra')
    plt.xlabel('Intensitas Pixel')
    plt.ylabel('Frekuensi')

    # Normalisasi histogram
    histogram_normalized = histogram / histogram.sum()

    # Menampilkan histogram yang sudah dinormalisasi
    plt.subplot(1, 3, 3)
    plt.plot(histogram_normalized, color='blue')
    plt.title('Histogram Normalisasi')
    plt.xlabel('Intensitas Pixel')
    plt.ylabel('Frekuensi Normalisasi')

    # Menampilkan semua hasil
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

**Penjelasan Source Code:**
1.  Membaca Citra
  - files.upload(): Digunakan untuk mengunggah file dari komputer lokal.
  - cv2.imread(image_path, cv2.IMREAD_GRAYSCALE): Membaca citra dalam format grayscale (hitam-putih), di mana setiap piksel memiliki nilai intensitas dari 0 (hitam) hingga 255 (putih).

2. Menampilkan Citra Grayscale
  - plt.imshow(image, cmap='gray'): Menampilkan citra dalam mode grayscale.
  - plt.title('Citra Grayscale'): Memberi judul pada subplot citra grayscale.

3. Menghitung Histogram
  - cv2.calcHist([image], [0], None, [256], [0, 256]): Menghitung histogram intensitas piksel dari citra.
  - Histogram menunjukkan distribusi jumlah piksel untuk setiap nilai intensitas (0 hingga 255).

4. Menampilkan Histogram
  - plt.plot(histogram, color='black'): Memvisualisasikan histogram dalam grafik garis warna hitam.
  Sumbu X mewakili intensitas piksel, dan sumbu Y menunjukkan jumlah piksel.

5. Normalisasi Histogram
  - histogram_normalized = histogram / histogram.sum(): Membagi setiap nilai dalam histogram dengan jumlah total piksel untuk mendapatkan nilai frekuensi relatif (normalisasi).

6. Menampilkan Histogram Normalisasi

  - plt.plot(histogram_normalized, color='blue'): Menampilkan histogram yang sudah dinormalisasi dalam grafik warna biru.
  Sumbu Y sekarang mewakili frekuensi relatif (probabilitas).

7. Mengatur Layout dan Menampilkan Semua
  - plt.tight_layout(): Mengatur agar seluruh grafik tidak saling bertumpuk.
  - plt.show(): Menampilkan semua subplot (citra, histogram, dan histogram normalisasi).

**Implementasi Histogram dan Normalisasi:**

1. Histogram
Histogram memberikan informasi distribusi intensitas piksel. Dalam aplikasi pengolahan citra, ini digunakan untuk:

  - Menilai pencahayaan citra (apakah terlalu gelap atau terang).
  - Mengidentifikasi kontras citra.
  - Melakukan segmentasi citra.

2. Normalisasi Histogram
Normalisasi histogram digunakan untuk:

  - Membandingkan distribusi intensitas antara citra yang berbeda.
  - Memproses citra untuk aplikasi tertentu seperti equalization (peningkatan kontras).

**Analisis Kualitas Citra Berdasarkan Histogram:**

- Histogram Citra:
Berdasarkan histogram (grafik hitam), distribusi intensitas tampak mencakup rentang nilai tertentu:

  - Ada dua puncak intensitas utama (disebut bimodal distribution), yang menunjukkan bahwa citra memiliki area gelap dan terang yang cukup kontras. Ini menunjukkan bahwa pencahayaan citra cukup baik.
- Normalisasi Histogram:
Nilai yang dinormalisasi menunjukkan frekuensi relatif intensitas piksel. Grafik histogram normalisasi memberikan pemahaman yang lebih jelas tentang proporsi piksel di setiap tingkat intensitas.

- Kategori Citra Baik:
Citra ini dapat dikategorikan baik, karena:

  - Histogram mencakup rentang nilai intensitas yang luas, dari area gelap hingga terang.
  - Ada kontras yang cukup antara bagian terang dan gelap, yang biasanya diinginkan dalam pengolahan citra.
  
Jika histogram hanya terpusat pada intensitas rendah atau tinggi saja, itu menunjukkan citra terlalu gelap atau terlalu terang, yang dapat menjadi indikator kualitas citra yang buruk.
