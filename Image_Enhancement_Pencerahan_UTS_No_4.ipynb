```python
from google.colab import files
uploaded = files.upload()  # Mengunggah file gambar

import cv2
import numpy as np
from matplotlib import pyplot as plt

def brighten_image(image, brightness_value):
    """
    Fungsi untuk mencerahkan citra dengan menambahkan konstanta ke setiap piksel
    dan melakukan operasi clipping agar nilai tetap dalam rentang 0-255.

    Args:
        image: numpy array, citra input dalam format grayscale atau RGB
        brightness_value: int, nilai konstanta untuk pencerahan

    Returns:
        brightened_image: numpy array, citra setelah pencerahan
    """
    # Pastikan nilai piksel tidak melebihi rentang [0, 255]
    brightened_image = cv2.add(image, brightness_value)
    return brightened_image

# Load gambar grayscale
image_path = next(iter(uploaded.keys()))
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Gambar tidak ditemukan!")
else:
    # Nilai konstanta untuk pencerahan
    brightness_value = 50  # Ubah nilai ini untuk tingkat pencerahan yang berbeda

    # Terapkan operasi pencerahan
    brightened_image = brighten_image(image, brightness_value)

    # Tampilkan gambar asli dan hasil pencerahan
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Gambar Asli")
    plt.imshow(image, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Gambar Setelah Pencerahan")
    plt.imshow(brightened_image, cmap='gray')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

**Penjelasan Operasi**
1. Pencerahan: Operasi ini dilakukan dengan menambahkan konstanta pada setiap nilai piksel. Contohnya, jika nilai piksel awal adalah 100 dan konstanta yang ditambahkan adalah 50, nilai piksel baru menjadi 150.

2. Operasi Clipping: Dalam representasi citra digital, nilai piksel harus berada dalam rentang [0, 255]. OpenCV menyediakan fungsi cv2.add yang secara otomatis menangani operasi clipping. Jika hasil penjumlahan melebihi 255, maka nilai akan diset ke 255. Jika kurang dari 0, nilai akan diset ke 0.

**Perubahan pada Citra Setelah Pencerahan**
1. Kecerahan Bertambah:
  - Dibandingkan dengan gambar asli (kiri), gambar setelah pencerahan (kanan) terlihat lebih terang.
  - Ini disebabkan oleh penambahan konstanta pada nilai intensitas piksel, yang menaikkan tingkat kecerahan secara keseluruhan.

2. Detail Area Gelap Lebih Jelas:
- Pada gambar asli, beberapa area tampak gelap, misalnya:
  - Latar belakang di belakang pembicara.
  - Bayangan pada pakaian pembicara.
- Setelah pencerahan, area gelap ini menjadi lebih terang sehingga detail pada area tersebut lebih terlihat.
3. Saturasi pada Area Terang:
- Beberapa bagian gambar, seperti layar proyektor yang sudah terang pada gambar asli, mengalami saturasi (mencapai nilai 255) pada gambar hasil pencerahan. Area ini tampak sepenuhnya putih, sehingga detail di area tersebut hilang.

4. Perubahan Kontras:
- Karena seluruh piksel bertambah secara merata, perbedaan antara piksel terang dan gelap sedikit berkurang. Hal ini dapat menyebabkan kontras dalam citra secara keseluruhan berkurang.

**Hasil Gambar**

Pada gambar hasil pencerahan:
- Kecerahan meningkat, membuat detail pada area gelap lebih terlihat.
- Beberapa area terang mengalami saturasi, yang menyebabkan hilangnya detail di area tersebut.
- Operasi clipping berhasil menjaga nilai piksel dalam rentang valid, memastikan hasil tidak melebihi nilai maksimum.

Jika diperlukan, pencerahan dapat disesuaikan dengan nilai
C yang lebih kecil untuk menghindari saturasi di area terang.
